```python
# exercise 1
number = int(input("Enter a number: ")) # ask user for a number
is_prime = True # assume number is prime

if number == 1: # 1 is not a prime number
    is_prime = False

elif number < 1: # numbers less than 1 are not prime
    is_prime = False

for i in range(2, number): # check divisors from 2 to number-1
    if number % i == 0: # if divisible with no remainder
        is_prime = False # number is not prime
        break # stop loop if divisor found

if is_prime: # if number prime
    print("This is a prime number") # print prime
else:
    print("This is not a prime number") # print not prime
    
```

```python
# exercise 2 
n = int(input("Enter how many Fibonacci numbers to display: ")) # ask how many Fibonacci numbers should be generated
a = 0 # the first Fibonacci number
b = 1 # the second Fibonacci number
print(a, end=" ") # print the first number and this function writes numbers in series and space between semi-columns, creates space between them
print(b, end=" ") # print the second number, also in series and have space 
for i in range(n - 2): # loop to generate the remaining Fibonacci numbers
    # calculate the next Fibonacci number
    c = a + b
    # print the next number in the sequence
    print(c, end=" ")
    # update the value of a to the previous b
    a = b
    # update the value of b to the new Fibonacci number
    b = c
```

```python
# exercise 3 
n = 5 # create variable, max number of star is 5 

for i in range(n, 0, -1): # cycle that decrease number of stars 
    for j in range(i): # loop to print stars in the row
        print('*', end=' ') # prints stars, dont go for next row and that there are spacwe
    print() # goes to next row 

for i in range(2, n + 1): # amount of stars increase 
    for j in range(i): # loop to print stars in the row
        print('*', end=' ') # prints star and space
    print() # goes to next row 
```

```python
# exercise 5 
def find_min_max(numbers): # define a function 
    min_value = min(numbers) # inside the function we use functions
    max_value = max(numbers)
    return min_value, max_value # return both values

nums = [8, 3, 2, 6, 4] # our numbers 

min_val, max_val = find_min_max(nums) # unpack results

print("Min value is:", min_val) # printing minimum value
print("Max value is:", max_val) # printing maximum value
```

```python
# exercise 6 
import random # import random numbers 
secret_number = random.randint(1, 10) # generates the number between 1 and 10 
guess = 0 # start variable and its value is 0 
while guess != secret_number: # start a loop 'while' that continues till the user guesses
    guess = int(input("Guess the number (1-10): ")) # ask to enter number and convert string to integer 

    if guess > secret_number: # check if the guessed number is bigger 
        print("Too high! Try again.") # printing that the number is bigger 

    elif guess < secret_number: # check if the guessed number is smaller 
        print("Too low! Try again.") # printing that the number is smaller

print("Congratulations! You guessed the number.") # print that the number was guessed
```

```python
# exercise 8
import math # import math to use factorial

def calculate_series(x, N): # define function to calculate the series
    total = 0 # variable to store the sum

    for i in range(1, N + 1): # loop through N elements of the series, i represents which term of the series we are calculating
        numerator = x * (5 ** (2 * i)) # calculate numerator 
        denominator = i + math.factorial(i + 1) # calculate denominator with factorial
        total += numerator / denominator # add current term to the total sum

    return total # return the final result

x = int(input("Enter value of x: ")) # ask user for x
N = int(input("Enter number of elements N: ")) # ask user for N

result = calculate_series(x, N) # call the function

print("Series result:", result) # print the result
```

```python
# exercise 9 
import numpy as np # import numpy
import matplotlib.pyplot as plt # import matplotlib
import PIL  # import image library
import urllib # import tool to load image from url

image_url = "https://cdn.wallpapersafari.com/30/62/jHBzTk.jpg" # image link

image_data = urllib.request.urlopen(image_url)
image = PIL.Image.open(image_data)
np_image = np.array(image)

horizontal = np.fliplr(np_image) # horizontal mirror
vertical = np.flipud(np_image) # vertical mirror

plt.subplot(1,3,1) # first image
plt.imshow(np_image) # show original
plt.title("Original") # title of the first image
plt.axis("off") # to not show axis

plt.subplot(1,3,2) # second image
plt.imshow(horizontal) # show horizontal mirror
plt.title("Horizontal") # title of the second image
plt.axis("off") # to not show axis

plt.subplot(1,3,3) # third image
plt.imshow(vertical) # show vertical mirror
plt.title("Vertical") # title of the third image
plt.axis("off") # to not show axis

plt.show() # display images
```

```python
# exercise 10
import numpy as np # import numpy

temps = [22,24,19,21,23,25,26,27,24,23,20,22,21,19,18,22,23,24,26,28,30,27,25,26,24,22,21,20,19,23] # temperature list

temp_array = np.array(temps) # convert list to numpy array

print("Min:", temp_array.min()) # minimum temperature
print("Max:", temp_array.max()) # maximum temperature
print("Average:", temp_array.mean()) # average temperature

fahrenheit = temp_array * 9/5 + 32 # convert celsius to fahrenheit
print("Fahrenheit:", fahrenheit)

above25 = temp_array[temp_array > 25] # temperatures above 25C
print("Above 25C:", above25)

below20 = np.sum(temp_array < 20) # count days below 20C
print("Days below 20C:", below20)

sorted_temp = np.sort(temp_array) # sort temperatures
print("Sorted:", sorted_temp)

reshaped = temp_array.reshape(3,10) # reshape to 5x6 array
print("Reshaped array:\n", reshaped)
```
